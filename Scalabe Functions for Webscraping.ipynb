```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen, urljoin
import os
import pathlib
import re
```

# Function to get all the `sublinks` from the main website

```python
def get_sublinks(url_main):
    
    ## This function takes link of the main page of a website and returns sublinks
    #  Input parameter: url of the website in quotes
    #  Output parameter: all the sublinks from the main page of a website 
    
    url  = (url_main)
    target_url  = (url_main)
    
    req = requests.get(url)
    
    soup = BeautifulSoup(req.text, "lxml")
    print(soup.title)
    print('')
    
    links = []
    
    for link in soup.find_all('a'):
        
        print(urljoin(target_url, link.get('href')))
        links.append(urljoin(target_url, link.get('href')))
    
    print(links)
```

```python
get_sublinks('https://www.junaidjamshed.com/')
```
